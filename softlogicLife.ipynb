```python
import psycopg2 as ps
import pandas as pd
import matplotlib.pyplot as plt, mpld3
import seaborn as sn
```

```python
conn = ps.connect(database="softlogic",
                        user='postgres', password='postgres', 
                        host='localhost', port='5432'
)
conn.autocommit = True
cursor = conn.cursor()
```

```python
cur = conn.cursor()

with open('facebookSoftlogicLife.csv', 'r') as f:
     
# Skip the header row.
    next(f)
    cur.copy_from(f, 'softlogiglifefb', sep=',')

conn.commit()
```

```python
df = pd.read_sql("SELECT * FROM softlogiglifefb", conn)
df
```
